```python
# Instalar as bibliotecas necessárias
!pip install --quiet gdown openpyxl

# Importar a biblioteca para baixar o arquivo
import gdown

# Definir o ID do arquivo no Google Drive
file_id = "10tpwe3_rfH-TZEuXpNT6vCUaqOK1aofZ"
# Criar a URL para acessar o arquivo
url = f"https://drive.google.com/uc?id={file_id}"
# Nome do arquivo que será salvo
output = "default_of_credit_card_clients.xls"

# Baixar o arquivo
gdown.download(url, output, quiet=False)

# Importar a biblioteca pandas para manipulação dos dados
import pandas as pd

# Definir manualmente os nomes das colunas
column_names = [
    'ID', 'LIMITE_CREDITO', 'SEXO', 'EDUCACAO', 'ESTADO_CIVIL', 'IDADE',
    'PAGAMENTO_SET', 'PAGAMENTO_AGO', 'PAGAMENTO_JUL', 'PAGAMENTO_JUN', 'PAGAMENTO_MAI', 'PAGAMENTO_ABR',
    'FATURA_SET', 'FATURA_AGO', 'FATURA_JUL', 'FATURA_JUN', 'FATURA_MAI', 'FATURA_ABR',
    'PAGTO_SET', 'PAGTO_AGO', 'PAGTO_JUL', 'PAGTO_JUN', 'PAGTO_MAI', 'PAGTO_ABR',
    'INADIMPLENTE_PROX_MES'
]

# Carregar o dataset e ajustar os nomes das colunas
df = pd.read_excel(output, header=1)
df.columns = column_names

# Selecionar as colunas que contêm as faturas mensais
fatura_features = ['FATURA_SET', 'FATURA_AGO', 'FATURA_JUL',
                   'FATURA_JUN', 'FATURA_MAI', 'FATURA_ABR']

# Obter as estatísticas descritivas das faturas
descricao = df[fatura_features].describe()

# Traduzir os índices das estatísticas para o português
descricao.index = [
    'Contagem', 'Média', 'Desvio Padrão', 'Mínimo',
    '1º Quartil (25%)', 'Mediana (50%)', '3º Quartil (75%)', 'Máximo'
]

# Exibir as estatísticas das faturas mensais
print("Estatísticas das faturas mensais (valores em NT$):")
print(descricao)



```

Exercício 3 – Histograma das faturas (2x3, 20 bins)

```python
import pandas as pd

# Recarregar o arquivo sem cabeçalho para garantir que tudo seja lido corretamente
df = pd.read_excel("default_of_credit_card_clients.xls", header=None)

# Mostrar as primeiras linhas para entender a estrutura dos dados
print("Primeiras linhas do arquivo (sem cabeçalho):")
print(df.head(10))

# Ajustar a primeira linha para ser usada como cabeçalho
df.columns = df.iloc[0]  # Definir a primeira linha como nomes das colunas
df = df.drop(0)  # Remover a primeira linha, que agora é o cabeçalho

# Limpar quaisquer espaços extras nos nomes das colunas
df.columns = df.columns.str.strip()

# Exibir as primeiras linhas agora com os cabeçalhos corrigidos
print("\nPrimeiras linhas do arquivo com cabeçalho corrigido:")
print(df.head(10))

# Filtrar as colunas que contêm 'BILL' e 'PAY' nos nomes
bill_features = [col for col in df.columns if 'BILL' in col.upper()]
pay_features = [col for col in df.columns if 'PAY' in col.upper()]

# Mostrar as colunas de fatura e pagamento encontradas
print("\nColunas de fatura (BILL):", bill_features)
print("Colunas de pagamento (PAY):", pay_features)

# Visualizar as faturas com histogramas
import matplotlib.pyplot as plt

# Configurar o gráfico
plt.figure(figsize=(15, 10))
for i, feature in enumerate(bill_features):
    plt.subplot(2, 3, i+1)  # Organizar os gráficos em uma grade
    df[feature].hist(bins=20)  # Criar o histograma para cada coluna
    plt.title(f'Histograma - {feature}')
    plt.xlabel('Valor da Fatura (NT$)')
    plt.ylabel('Frequência')

# Ajustar o layout para que os gráficos não fiquem um em cima do outro
plt.tight_layout()
plt.show()





```

Exercício 5 – Histogramas dos pagamentos com rotação nos rótulos

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Carregar o arquivo de dados
df = pd.read_excel("default_of_credit_card_clients.xls", header=0)

# Filtrar as colunas relacionadas ao pagamento (PAY)
pay_features = [col for col in df.columns if 'PAY' in col.upper()]

# Verificar os tipos de dados das colunas de pagamento
print("Tipos de dados nas colunas de pagamento:")
print(df[pay_features].dtypes)

# Tentar converter as colunas de pagamento para valores numéricos
df[pay_features] = df[pay_features].apply(pd.to_numeric, errors='coerce')

# Remover linhas com valores ausentes nas colunas de pagamento
df_cleaned = df.dropna(subset=pay_features)

# Calcular o número de subgráficos (subplots) necessários
num_plots = len(pay_features)
num_rows = math.ceil(num_plots / 3)  # Calcular o número de linhas para os gráficos
num_cols = 3  # Definir 3 colunas por linha

# Plotar histogramas para cada coluna de pagamento
plt.figure(figsize=(15, 5 * num_rows))  # Ajustar o tamanho da figura com base no número de linhas
for i, feature in enumerate(pay_features):
    plt.subplot(num_rows, num_cols, i + 1)  # Organizar os subgráficos
    df_cleaned[feature].hist(bins=20)
    plt.title(f'Histograma - {feature}')
    plt.xlabel('Valor do Pagamento')
    plt.ylabel('Frequência')

# Ajustar layout e exibir os gráficos
plt.tight_layout()
plt.show()


```

Exercício 7 – Transformação logarítmica (pagamentos ≠ 0)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# Carregar os dados do arquivo Excel
df = pd.read_excel("default_of_credit_card_clients.xls", header=0)

# Filtrar as colunas que têm "PAY" no nome
pay_features = [col for col in df.columns if 'PAY' in col.upper()]

# Verificar os tipos de dados dessas colunas de pagamento
print("Tipos de dados nas colunas de pagamento:")
print(df[pay_features].dtypes)

# Tentar converter essas colunas para números, caso tenha algum texto ou erro
df[pay_features] = df[pay_features].apply(pd.to_numeric, errors='coerce')

# Criar uma "máscara" para verificar onde o pagamento é igual a zero
pay_zero_mask = df[pay_features].eq(0)

# Contar quantos pagamentos são iguais a zero
print("Contagem de pagamentos iguais a zero:")
print(pay_zero_mask.sum())

# Ignorar os pagamentos iguais a zero para as análises seguintes
df_non_zero_payments = df[~pay_zero_mask]

# Aplicar o logaritmo base 10 nos pagamentos que não são zero
df_log_payments = df_non_zero_payments[pay_features].apply(np.log10)

# Calcular quantos gráficos (subplots) vamos precisar
num_plots = len(pay_features)
num_rows = math.ceil(num_plots / 3)  # Calcular quantas linhas de gráficos
num_cols = 3  # Definir 3 gráficos por linha

# Plotar os histogramas dos pagamentos transformados pelo logaritmo
plt.figure(figsize=(15, 5 * num_rows))  # Ajustar o tamanho da figura com base nas linhas
for i, feature in enumerate(pay_features):
    plt.subplot(num_rows, num_cols, i + 1)  # Organizar os gráficos
    if feature in df_log_payments.columns:  # Verificar se a coluna existe
        df_log_payments[feature].hist(bins=20)
        plt.title(f'Log10 de {feature}')
        plt.xlabel('Log10(Pagamento)')
        plt.ylabel('Frequência')

# Ajustar o layout para que os gráficos fiquem bem organizados
plt.tight_layout()
plt.show()

```
